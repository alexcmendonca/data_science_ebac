````python
# Criando uma função em Python que recebe como argumento o UF da consulta e o tipo de interessado ('Estado'ou Município) em um DataFrame.
def consulta_dataframe_uf(uf:str, tipo_interessado:str):
    df_uf = df_base[df_base['uf'] == uf]
    if tipo_interessado == 'Estado':
        df_function = df_uf[df_uf['tipo_interessado'] == 'Estado']
    elif tipo_interessado == 'Município':
        df_function = df_uf[df_uf['tipo_interessado'] == 'Município']
    return df_function
````
