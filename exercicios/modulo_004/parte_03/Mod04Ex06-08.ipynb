````python
# validando informações disponíveis na variável
df_base['tipo_interessado'].unique()
````

````python
# criando função para filtrar dados no dataframe
def filtro_dataframe(uf, tipo_interessado):
    # verificando se o tipo_interessado é um valor válido
    informacoes_validas = ['Estado', 'Município']
    if tipo_interessado not in informacoes_validas:
        raise ValueError(f'Informar valores possíveis: {informacoes_validas}')
    
    # Filtra o DataFrame pelo uf
    df_uf = df_base[df_base['uf'] == uf]

    # Filtra o df pelo tipo_interessado
    df_function = df_uf[df_uf['tipo_interessado'] == tipo_interessado]

    return df_function
````

````python
# Utilizando a função para consultar info da uf = MG e tipo_interessado = Estado
filtro_dataframe('MG', 'Município')
````

````python
# Verificando quantas solicitações para o Estado podem ser consultadas para Minas Gerais com status em 'Arquivado por decurso de prazo'.
df_uf = df_base[df_base['uf'] == 'MG']
df_status = df_uf['status'] == 'Arquivado por decurso de prazo'
df_status.sum()
````
