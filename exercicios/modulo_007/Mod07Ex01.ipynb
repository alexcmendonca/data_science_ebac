#### 2) Vamos montar um metadados

1. Crie um dataframe com os nomes de cada variável e o tipo de dados de cada variável.
2. Adicione uma coluna nesse *dataframe* chamada "qtd_categorias" e coloque nela o número de categorias correspondente de cada variável. 
    Dica: 
        1. inicie uma lista vazia
        2. faça um for ao longo dos nomes das variáveis, 
        3. conte o número de categorias dessa variável 
        4. acumule essa informação de 3. na lista que você criou em 1. 
        5. No final, essa lista pode ser atribuída à nossa variável.
3. Crie variáveis dummy para as variáveis necessárias (i.e. aquelas que são qualitativas e não estão armazenadas como {0, 1} ou {True, False}.

````python
# Analisando os dados
df.dtypes
````

````python
# Analisando os dados
df.columns
````

````python
# Criando um dataframe com os nomes de cada variável e o tipo de dados de cada variável
df_variaveis = pd.DataFrame({'variaveis': df.columns, 'tipo_dado': df.dtypes})
df_variaveis.reset_index(drop = True, inplace = True)
df_variaveis
````

````python
# Consultando variável com o qtde de categorias disponíveis
df.tipo_renda.unique()
````

````python
df.tipo_renda.value_counts()
````

````python
df['sexo'].dtype
````

````python
# Criando uma lista vazia para salvar o número de categorias
categorias = []

# Criando um for ao longo dos nomes das variáveis
for var in df_variaveis['variaveis']:
    # Consultando o tipo da variável
    tipo_variavel = df[var].dtype
    if tipo_variavel == 'object':
        # Contando o número de categorias de cada variável e acumulando as informações
        categorias_tamanho = len(df[var].unique())
        categorias.append(categorias_tamanho)
    else:
        # Atribuindo valor false = 0 para os demais registros
        categorias.append(0)

# Atribuindo os valores na nova coluna chamada de 'qtd_categorias'
df_variaveis['qtd_categorias'] = categorias
````

````python
df_variaveis
````

````python
# Validando a qtde de categorias disponíveis na variável
df.tipo_residencia.unique()
````

````python
# Analisando df com variáveis que possuem qtd_categorias maior que 1
df_var_maior_categorias = df_variaveis[df_variaveis['qtd_categorias'] > 1]
df_var_maior_categorias
````

````python
# Analisando df para criar variáveis em dummie, mapping, int
df[['sexo', 'posse_de_veiculo', 'posse_de_imovel', 'tipo_renda', 'educacao', 'estado_civil', 'tipo_residencia']].head()
````

````python
# Alterando tipo da variável utilizando Mapping
df.sexo = df.sexo.map({'F': 1, 'M': 0})
````

````python
# Alterando tipo da variável utilizando Mapping
df.posse_de_veiculo = df.posse_de_veiculo.map({'Y': 1, 'N': 0})
````

````python
# Alterando tipo da variável utilizando Mapping
df.posse_de_imovel = df.posse_de_imovel.map({'Y': 1, 'N': 0})
````

````python
df[['sexo', 'posse_de_veiculo', 'posse_de_imovel', 'tipo_renda', 'educacao', 'estado_civil', 'tipo_residencia']].head()
````

````python
# Alterando tipo da variável com dummie
df_encoded = pd.get_dummies(df, columns=['tipo_renda', 'educacao', 'estado_civil', 'tipo_residencia'], drop_first=True)
df_encoded.head()
````

Considere as variáveis ```possui_email``` e ```posse_de_veiculo```. Faça uma tabela cruzada entre elas e responda qual delas te parece mais poderosa para prever a probabilidade de ```mau = 1```?
